```python
X=df.iloc[:, :-1] #represents the data upto a specific integer location
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
y=df.iloc[:,-1]

# Encode label category
# male -> 1
# female -> 0
#encodes target values with 0 or 1
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
y
```

```python
# Scale the data to be between -1 and 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #Standardize features by removing the mean and scaling to unit variance
scaler.fit(X)
X = scaler.transform(X) #performs standardisation by scaling
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) #splits the matrices and arrays into train and test sets
```

```python
from sklearn.svm import SVC #support vector clasification
from sklearn import metrics
svc=SVC() #Default hyperparameters
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
svc=SVC(kernel='rbf') #radial basis function finds the eucledian distance between 2 points divided by the sigmoid function
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
svc=SVC(kernel='poly')#It represents the similarity of feature vectors predicted along with the original data
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
svc=SVC(kernel='linear')
scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy') #cv is cross validation
print(scores)
```

```python
C_range=list(range(1,26))
acc_score=[]
for c in C_range:
    svc = SVC(kernel='linear', C=c)
    scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)  
```

```python
import matplotlib.pyplot as plt #SHOWS THE CURRENT TICK LOCATIONS AND LABELS
%matplotlib inline


C_values=list(range(1,26))
# plot the value of C for SVM (x-axis) versus the cross-validated accuracy (y-axis)
plt.plot(C_values,acc_score)
plt.xticks(np.arange(0,27,2)) #arange returns values at specified intervals.xticks sets the current ticks locations f x-axis
plt.xlabel('Value of C for SVC')
plt.ylabel('Cross-Validated Accuracy')
```
